## Análise de Eficiência das escolas

Para determinar a eficiencia dos diferentes tipos de escolas, será aplicado um teste de hipótese. 
O teste RankSums, também conhecido como teste de Soma de Postos de Wilcoxon-Mann-Whitney, é um teste que vai comparar dois grupos para verificar se eles provém de uma mesma distribuição de probabilidade. 

Esse teste será aplicado para as distribuições da nota normalizada no exame de cada escola (gênero masculino, gênero feminino e gênero misto).

### Hipótese I

1. A distribuição da nota normalizada no exame é diferente em escolas de gênero único (somente feminino ou somente masculino).

**H<sup>null</sup>**
>Não existe diferença na performance média dos alunos provenientes da escola de gênero masculino e da escola de gênero feminino.

**H<sup>alt</sup>**
>A performance média dos alunos provenientes da escola de gênero feminino é superior aos alunos provenientes da escola de gênero masculino.

```python
r_gender = ranksums(girls_df['normexam'], boys_df['normexam'])
print('O valor do p-value é {}'.format(r_gender.pvalue))
```

**H<sup>null</sup>**
>Não existe diferença na performance média entre os alunos das escolas de gênero feminino e masculino.

### Hipótese II

2. A distribuição da nota normalizada no exame é diferente em escolas de gênero somente **feminino** e **gênero misto**.

**H<sup>null</sup>**
>Não existe diferença na performance média dos alunos provenientes da escola de gênero feminino e da escola de gênero misto.

**H<sup>alt</sup>**
>A performance média dos alunos provenientes da escola de gênero feminino é superior aos alunos provenientes da escola de gênero misto.

```python
r_gender = ranksums(girls_df['normexam'], mixed_df['normexam'])
print('O valor do p-value é {}'.format(r_gender.pvalue))
```

**H<sup>alt</sup>**
>A performance média dos alunos provenientes da escola de gênero feminino é superior aos alunos provenientes da escola de gênero misto.

### Hipótese III

3. A distribuição da nota normalizada no exame é diferente em escolas de gênero somente **masculino** e **gênero misto**.

**H<sup>null</sup>**
>Não existe diferença na performance média dos alunos provenientes da escola de gênero masculino e da escola de gênero misto.

**H<sup>alt</sup>**
>A performance média dos alunos provenientes da escola de gênero feminino é superior aos alunos provenientes da escola de gênero misto.

```python
r_gender = ranksums(boys_df['normexam'], mixed_df['normexam'])
print('O valor do p-value é {}'.format(r_gender.pvalue))
```

**H<sup>null</sup>**
>Não existe diferença na performance média dos alunos provenientes da escola de gênero masculino e da escola de gênero misto.
